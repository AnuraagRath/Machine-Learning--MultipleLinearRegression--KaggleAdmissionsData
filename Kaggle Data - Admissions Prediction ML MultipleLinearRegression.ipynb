```python
import seaborn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import pandas as pd
import numpy as np
```

```python
admissionData = pd.read_csv("new.csv")
admissionData.head()
```

```python
plt.scatter(admissionData["GRE Score"], admissionData.Admit)
plt.xlabel('GRE Score')
plt.ylabel('CHANCE OF ADMISSION')
plt.show()
```

```python
plt.scatter(admissionData["CGPA"], admissionData.Admit, color = 'red')
plt.xlabel('CGPA')
plt.ylabel('CHANCE OF ADMISSION')
plt.show()
```

```python
plt.scatter(admissionData["University Rating"], admissionData.Admit, color = "purple")
plt.xlabel('University Rating')
plt.ylabel('CHANCE OF ADMISSION')
plt.show()
```

```python
plt.scatter(admissionData["TOEFL Score"], admissionData.Admit, color = "green")
plt.xlabel('TOEFL Score')
plt.ylabel('CHANCE OF ADMISSION')
plt.show()
```

```python
plt.scatter(admissionData.LetterOfRecommendation, admissionData.Admit, color = "violet")
plt.xlabel('Letter of Recommendation')
plt.ylabel('CHANCE OF ADMISSION')
plt.show()
```

```python
plt.scatter(admissionData.Research, admissionData.Admit, color = "orange")
plt.xlabel('Research')
plt.ylabel('CHANCE OF ADMISSION')
plt.show()
```

```python
x = admissionData[['GRE Score', 
                   'TOEFL Score', 
                   'University Rating', 
                   'SOP', 
                   'LetterOfRecommendation', 
                   'CGPA', 
                   'Research']]
y = admissionData.Admit
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, 
                                                y, 
                                                train_size = 0.8, 
                                                test_size = 0.2, 
                                                random_state=6)

admissionRegression = LinearRegression()
admissionRegression.fit(xTrain, yTrain)
predictAdmission = admissionRegression.predict(xTest)
```

```python
theScore = admissionRegression.score(xTrain, yTrain)
print("Accuracy of Model:", theScore)
```
